# **Data cleaning**

```python
import pandas as pd
```

```python
all_data = pd.read_csv('all_data.csv',index_col=0)
```

```python
all_data.head()
```

```python
all_data['name'] = all_data['name'].str.replace('Смартфон ', '', regex=False)
#delete the 'Смартфон'
```

```python
all_data.head()
```

```python
all_data['name'] = all_data['name'].str.replace(r'\d+ ГБ', '', regex=True)
#also delete the specifications cause we have the it in the other columns
```

```python
all_data['name'] = all_data['name'].str.replace('/', '', regex=True)
colors = ['starlight', 'бежевый', 'белый', 'бордовый', 'бронзовый', 'голубой', 
          'желтый', 'зеленый', 'золотистый', 'коричневый', 'красный', 'лавандовый', 
          'оранжевый', 'розовый', 'серебристый', 'серый', 'синий', 'сиреневый', 
          'темно-зеленый', 'темно-синий', 'фиолетовый', 'черный']

for color in colors:
    all_data['name'] = all_data['name'].str.replace(color, '', regex=False)
#take all colors in the kaspi filters and the delete it
```

```python
all_data['name']
```

```python
all_data['NFC'] = all_data['NFC'].map({'Да': 1, 'Нет': 0})
#map the NFC column, if smartphone have the NFC it's maybe higher, so NFC yes to 1, for greater significance
```

```python
all_data.head()
```

```python
median_reviews = all_data['reviews'].mean()

all_data['reviews'] = all_data.apply(lambda row: 0 if pd.isna(row['reviews']) and row['rating'] == 0 else row['reviews'], axis=1) 
#some phones have the 0 rating so it means that reviews 0
all_data['reviews'] = all_data['reviews'].fillna(median_reviews) 
#the moments when reviews it's not scrap right change it to median
```

```python
all_data['reviews'].value_counts()
```

```python
all_data['Диагональ'] = all_data['Диагональ'].str.replace(' дюйм', '').astype(float)
all_data['RAM'] = all_data['RAM'].str.replace(' ГБ', '').astype(float)
all_data['Память'] = all_data['Память'].str.replace(' ГБ', '').astype(float)
all_data['Емкость акк'] = all_data['Емкость акк'].str.replace(' мАч', '').astype(float)
#delete the words in this columns and convert it to numerical
```

```python
all_data
```

```python
all_data.info()
```

```python

all_data['price'] = all_data['price'].str.replace(' ₸', '').str.replace('\xa0', '').str.replace(' ', '').astype(int)
#price to numerical
```

```python
all_data.head()
```

```python
all_data['Тип экрана'].value_counts()
```

```python
important_screen_types = ['IPS', 'OLED', 'AMOLED']
#we have only 3 the main screen type
def clean_screen_type(screen_type):
    if 'IPS' in screen_type:
        return 'IPS'
    if 'AMOLED' in screen_type:
        return 'AMOLED'
    elif 'OLED' in screen_type:
        return 'OLED'
    else:
        return 'Other'  
all_data['Тип экрана'] = all_data['Тип экрана'].apply(lambda x: clean_screen_type(x))
#so change it
```

```python
all_data.head()
```

```python
all_data['Тип экрана'].value_counts()
```

```python
all_data['OS'] = all_data['name'].str.contains('Iphone', case=False).apply(lambda x: 'iOS' if x else 'Android')
#add the OS, cause the iOS have the higher price with equal specifications
```

```python
all_data = all_data[all_data['name'].apply(lambda x: '+' not in str(x))]
all_data= all_data.reset_index(drop=True)
#We have the phones with '+gift'
#this data is very disturbing because it does not show the real price, so i decided to drop it
```
